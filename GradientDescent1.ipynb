```python
import numpy as np
from sklearn.datasets import make_regression
```

```python
X,y=make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red')
```

```python
#calculating the Gradient Descent assuming slope is constant m=78.36
#assuming b=0 as of now
y_pred=((78.36*X)+0).reshape(4)
```

```python
y_pred
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='green',label='b=0')
plt.legend()
plt.show()
```

```python
#calculating slope at b=0
m=78.65
b=0
slope=-2*np.sum(y-m*X.ravel()-b)
slope
```

```python
#calculating learning rate 0.1
lr=0.1
step_size=slope*lr
step_size
```

```python
#calculating the new intercept
b=b-step_size
```

```python
b
```

```python
y_pred1=((78.36*X)+b).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='green',label='b=0')
plt.plot(X,y_pred1,color='blue',label='b={}'.format(b))
plt.legend()
plt.show()
```
